```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([122,126,132,142,153,162]).reshape(-1,1)
y = np.array([301,351,426,406,497,518])
plt.scatter(x,y)
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

```python
Ir = LinearRegression()
Ir.fit(x,y)
```

```python
w = Ir.coef_[0]
b = Ir.intercept_
print('slope: ', w)
print('intercept: ', b)
```

```python
plt.scatter(x,y)
# plt.plot(x, Ir.predict(x), color='red')
plt.xlabel("Area")
plt.ylabel("Price")
plt.plot([x[0], x[-1]], [x[0]*w+b,x[-1]*w+b])
plt.show()
```

```python
testX =np.array([125]).reshape(-1,1)
print(Ir.predict(testX))
```
